```python
with open("Example_Regex.txt", 'r', encoding='utf16') as file:
    text  = file.read() 
```

```python
import re 
```

```python
re.findall('[abc]', 'abcs dsf ádfa')
```

```python
#get all phone numbers
pattern = r'[0-9]{5}.[0-9]{3}.[0-9]{3}'
pattern1 = r'[0][0-9]{9}'
print(re.findall(pattern, text))
print(re.findall(pattern1, text))
#print(text)
```

```python
print(text)
```

```python
# get all dates
pattern2 = r'[0-9]{2}/[0-9]{2}/[0-9]{4}'
print(re.findall(pattern2, text))
```

```python
pattern = r'[0-9]{5}.[0-9]{3}.[0-9]{3}|[0][0-9]{9}'
print(re.findall(pattern, text))
```

```python
re.findall('(a|b)c','abc afb abc ac ')
```

```python
pattern3 = r'\w+@\w+.\w+.\w+'
re.findall(pattern3, text)
```

```python
i = 0
for line in text.split('\n'):
    result = re.search(pattern3, line)
    if result:
        print(result.group())
    else:
        i+=1
print(i)
```

```python
re.split('[?.!;]',text)
```

```python
print(re.sub(pattern3, '[Email]', text))
```
