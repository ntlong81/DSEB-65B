```python
def aveg_scores(name, *scores):
    return {'name':name, 
            "aveg_score":sum(scores)/len(scores)}
```

```python
aveg_scores('Lan',6,7,8,4,3,5)
```

```python
lan_score = [3,4,5,6,7,9,8,7,6]
aveg_scores('Lan',*lan_score)
```

```python
def gpa_scores(name,**scores):
    aveg = sum(scores.values())/len(scores)
    return {name:scores,"aveg":aveg}
```

```python
gpa_scores('Mai',Algebra = 8, 
           statistics = 9, 
           Python4DS =10,
           optimize = 8)
```

```python
Mai_score = {'algebra':9,'statistics':8}
gpa_scores("Mai",**Mai_score)
```

```python
def gen_number1(n):
    i = 0
    while i<n:
        return i
        i += 1
def gen_number2(n):
    i = 0
    while i<n:
        yield i
        i += 1
```

```python
def is_prime(n):
    if n<=1:
        return False 
    else:
        for d in range(2, int(n**0.5)+1):
            if n%d==0:
                return False 
    return True
```

```python
numbers = list(range(100,1000))
list(map(is_prime,numbers))
```

```python
#count the number of even number in each items
alist = [[1,2,3],[2,2,2],[2,3,4,6], list(range(2,10))]
list(map(sum,
         (map(lambda x: x%2==0,
              sublist) for sublist in alist)))
```

```python
# sum of all prime numbers in each items
[sum(list(map(lambda x: x * is_prime(x), lis))) for lis in alist]
```

```python
letters = list('sfafafjlewroiwen,c,laljfelsajfdalsjfd')
list(filter(lambda x: 
            x not in list('aieuo'), 
            letters))
```

```python
#sum all prime numbers in each items by filter function
list(map(sum,(filter(is_prime,item) for item in alist)))
```
